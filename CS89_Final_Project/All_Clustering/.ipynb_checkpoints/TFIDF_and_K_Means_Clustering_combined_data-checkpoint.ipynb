```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN

# Insspiration for tsne and top-k words: https://www.kaggle.com/code/jbencina/clustering-documents-with-tfidf-and-kmeans
```

```python
df = pd.read_csv('combined_text_data.csv')
df['Date'] = pd.to_datetime(df['Date']).astype('int64')
df
```

```python
# Keeping the date, type, author, and text
df = df.drop(['Unnamed: 0'], axis=1)
df.head(10)
```

```python
# Give each author an authorId
temp = 0
author_dict = {}

#First we get each unique author and assign them a numerical key authorId
for author in df['Author']:
    if author in author_dict:
        author = author_dict[author]
    else:
        author_dict[author] = temp
        temp+=1
#Populate the DF with respective authors
df['authorId'] = [author_dict[author] for author in df['Author']]

#NOTE: This keeps all parts of the author's name within the text including their title
# We chose not to remove this as their may be importance in the value of a given persons speeches before
# vs after appointment to a certain position. (i.e. Vice Chairman --> Chairman)
```

```python
#Setting TFIDF Parameters

tfidf = TfidfVectorizer(
    min_df = 5,
    max_df = 0.8,
    max_features = 8000,
    stop_words = 'english'
)
tfidf.fit(df.Text)
text = tfidf.transform(df.Text)
```

```python
#Elbow method using WCSS

X = text.copy()
wcss = {} 
for i in range(1, 20): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 20, n_init = 'auto')
    kmeans.fit(X)
    wcss[i] = kmeans.inertia_
     
plt.plot(wcss.keys(), wcss.values(), 'gs-')
plt.xlabel("Values of 'k'")
plt.ylabel('WCSS')
plt.show()
```

```python
#Plot PCA and TSNE Function

def plot_tsne_pca(data, labels, n):

    max_label = max(labels)
    
    # PCA
    pca = PCA(n_components=2).fit_transform(np.asarray(data.todense()))

    # t-SNE
    tsne = TSNE().fit_transform(PCA(n_components=50).fit_transform(np.asarray(data.todense())))
    
    f, ax = plt.subplots(1, 2, figsize=(14, 6))
    
    ax[0].scatter(pca[:, 0], pca[:, 1], c=labels)
    ax[0].set_title('PCA Cluster Plot of Speeches, Statements, and Minutes. '+str(n)+' clusters.')
    
    ax[1].scatter(tsne[:, 0], tsne[:, 1], c=labels)
    ax[1].set_title('t-SNE Cluster Plot of Speeches, Statements, and Minutes. '+str(n)+' clusters.')
    
```

```python
#Plotting 4 clusters

clusters = KMeans(n_clusters=4, random_state=20).fit_predict(text)
plot_tsne_pca(text, clusters, 4)
```

```python
#Getting top 10 key words

def get_top_keywords(data, clusters, labels, n_terms):
    df = pd.DataFrame(data.todense()).groupby(clusters).mean()
    
    for i,r in df.iterrows():
        print('\nCluster {}'.format(i))
        print(','.join([labels[t] for t in np.argsort(r)[-n_terms:]]))
            
get_top_keywords(text, clusters, tfidf.get_feature_names_out(), 10)
```
