```python
import pandas as pd
import datetime as dt
import time
```

```python
#Import speeches DF
# speeches_df = pd.read_csv('../All_Scraping/df_2000_to_2203.csv')
speeches_df = pd.read_csv('../Text_Scraping/df_2000_to_2203.csv')
speeches_df = speeches_df[['Date', 'Author', 'Text']]
speeches_df['Type'] = 'Speech'

#Import Statements/Minutes DF
# fomc_statements_minutes_df = pd.read_csv('../All_Scraping/FOMC_Statements_and_Minutes.csv')
fomc_statements_minutes_df = pd.read_csv('../Text_Scraping/FOMC_Statements_and_Minutes.csv')
fomc_statements_minutes_df = fomc_statements_minutes_df[['Date', 'Type', 'Text']]
#Convert type column to author
fomc_statements_minutes_df['Author'] = ['FOMC '+ t for t in fomc_statements_minutes_df['Type']]
# fomc_statements_minutes_df.rename(columns = {'Type':'Author'}, inplace = True)

#Merge the two dataframes
all_text_df = pd.concat([speeches_df, fomc_statements_minutes_df])
#Convert date to datetime
all_text_df['Date'] = pd.to_datetime(speeches_df['Date'])

display(all_text_df.head(10))
print('-----')
display(all_text_df.tail(10))
print('-----')
print(all_text_df.shape)
```

```python
 all_text_df.to_csv('combined_text_data.csv')
```
